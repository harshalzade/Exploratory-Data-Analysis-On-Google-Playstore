### Category

In category column 1.9 is wrong value ,so i have to see correspond values of 1.9

so we have change value of  category 1.9 into  the nan

```python
df.loc[10472,'Category'] = np.nan
```

### Rating

```python
df['Rating'].unique()
```

```python
df.Rating.dtype
```

### Reviews

```python
df['Reviews'].unique()
```

```python
df[df['Reviews']=='3.0M']
```

```python
df['Reviews']= df.Reviews.replace('3.0M','3000000.0')
df['Reviews']= df.Reviews.replace('0.0','0')
```

```python
df['Reviews'] = df.Reviews.astype(float)
```

```python
df.loc[10470:10474]
```

### Size

```python
df['Size'].unique()
```

Here we Convert M(Megabyte) into 1000000 byte and k(kilobyte) into 1000 byte

'Varies with device' into nan

 and convert value 1000+ into 1000

```python
df['Size'] = df['Size'].str.replace('M','000000')
df['Size'] = df['Size'].str.replace('k','000')
df['Size'] = df['Size'].replace('1,000+',1000)
df['Size'] = df['Size'].replace('Varies with device',np.nan)
df['Size'] = df['Size'].astype(float)
```

```python
df['Size'].dtype
```

### Installs


```python
df['Installs'].unique()
```

We can convert str datatype into float and Doing Same process like Size

```python
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')
df['Installs'] = df['Installs'].replace('Free',np.nan)
df['Installs'] = df['Installs'].astype(float)
```

```python
df['Installs'].dtype 
```

### Type

```python
df['Type'].unique()
```

### Price

```python
df['Price'].unique()
```

```python
df['Price'] = df['Price'].str.replace('$','')
df['Price'] = df['Price'].replace('Everyone',np.nan)
df['Price'] = df['Price'].astype(float)
```

Data type of Price is object. I have made similar processes, which I made the in 'Size'.

```python
df['Price'].unique()
```

```python
df.head(2)
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
df['Last Updated']
```

There seems to be a problem with the Rating column may contain invalid values due to accidental or intentional errors while responding. A simple fix would be ignore the rows where the value in the Rating column is higher than 5  This can be done using the .drop method, as explained here.

```python
 df.drop(df[df.Rating > 5].index,inplace=True)
```

```python
df.describe()
```

You can check the total number of missing values in each column with a single line of code.

```python
df.isnull().sum()
```

You could also check the missing values as a percent of all observations.

```python
df.isnull().sum()*100/len(df)
```

 **One treatment of missing values is just dropping the observations entirely if the dataset is relatively large and missing cases are few.**

 **Dropping is only advised to be used if missing values are few `(say 0.01â€“0.5% of our data)`. Percent is just a rule of thumb.**

#####  In our datasets `Type`,`Current ver`,`Android ver` lies between `(say 0.01â€“0.5% of our data)` so we drop nan values


```python
df.dropna(subset=['Type'],inplace=True)
df.dropna(subset=['Current Ver'],inplace=True)
df.dropna(subset=['Android Ver'],inplace=True)
```

```python
df.isnull().sum()
```

## Exploratory Analysis and Visualization

```python
import matplotlib
import sys
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

corr() : It returns correlation.
    
describe (): It returns number of entries, average of entries, outlier values, standart deviation, minimum and maximum entry.

```python
#correlation map
f,ax = plt.subplots(figsize=(12, 12))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

**We can identify the `Application` with the TOP `10` highest number of respondents using the value_counts method.**

```python
s = df.App.value_counts().head(10)
s
```

```python
names = list(s.index)
values = list(s.values)
    

fig,ax = plt.subplots(1, 3, figsize=(20, 8), sharey=True)


ax[0].bar(names, values)
ax[1].scatter(names, values)
ax[2].plot(names, values)

fig.autofmt_xdate(rotation=75)

fig.suptitle('TOP 10 App highest number of respondents')
plt.show()
```

**Now we find count of Rating Per Category and Total Review per Category**

```python
categ_rating = df.groupby('Category').Rating.nunique()
categ_rev = df.groupby('Category').Reviews.sum()
```

```python
# Set up 2 x 1 chart group
figCategory, chtCategory = plt.subplots(2, 1, figsize=(20, 15))

# Set up x-axis for 2 charts
srCategoryList = categ_rating.index

# Use the cht for plotting
chtCategory[0].plot(srCategoryList, categ_rating.values, marker='s')
chtCategory[0].set_xlabel('Category')
chtCategory[0].set_ylabel('Rating Count')
chtCategory[0].set_title(' Count Of Rating per Category')
chtCategory[0].set_xticklabels(srCategoryList, rotation=75)


chtCategory[1].plot(srCategoryList, categ_rev.values, marker='s')
chtCategory[1].set_xlabel('Category')
chtCategory[1].set_ylabel('Total Reviews')
chtCategory[1].set_title('Total Reviews per Category')
chtCategory[1].set_xticklabels(srCategoryList, rotation=75)

plt.tight_layout(pad=5)
plt.show()
```

**Heat maps display numeric tabular data where the cells are colored depending upon the contained value. Heat maps are great for making trends in this kind of data more readily apparent, particularly when the data is ordered and there is clustering.**

Here we comapre `Genres` and `Content Rating` with values of `Size`

`For Better View please Zoom the heatmap`

```python
plt.rcParams['figure.figsize'] = (50.0, 80.0)
plt.rcParams['font.size'] = 28
bg_color = (0.88,0.85,0.95)
plt.rcParams['figure.facecolor'] = bg_color
plt.rcParams['axes.facecolor'] = bg_color
fig, ax = plt.subplots(1)
df_s = pd.pivot_table(data=df,
                    index='Genres',
                    values='Size',
                    columns='Content Rating')
p = sns.heatmap(df_s, cmap='coolwarm',annot=True,
                fmt=".1f",
                annot_kws={'size':25},
                ax=ax)

plt.text(5,30.3, "Heat Map", fontsize = 15, color='Black', fontstyle='italic')
plt.show()

print("For Better View please Zoom the heatmap")
```
