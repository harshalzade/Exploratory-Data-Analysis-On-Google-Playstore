```python
cols = ['Quantity','UnitPrice']
```

```python
Sale = csvdata[cols]
```

```python
Sale.head(11)
```

```python
Sale['Revenue'] = Sale['Quantity'] * Sale['UnitPrice']*75
```

```python
Sale.head(11)
```

```python
Sale.tail(11)
```

```python
Sale.info()
```

```python
Sale['RevenueFlag'] = Sale['Revenue'] > Sale['Revenue'].mean()
```

```python
Sale.head(11)
```

```python
Sale['Revenue'].mean()
```

```python
totalrevenue = sum(Sale['Revenue'])
```

```python
totalrevenue
```

```python
num_of_obs = Sale['Revenue'].shape[0]
```

```python
num_of_obs
```

```python
AverageRevenue = totalrevenue/num_of_obs
```

```python
AverageRevenue
```

```python
revenue = Sale['Revenue']
```

```python
cheack = []
```

```python
for value in revenue:
    if(value>AverageRevenue):
        cheack.append(1)
    else:
        cheack.append(0)
```

```python
Sale['AboveAverage'] = cheack
```
